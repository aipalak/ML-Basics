```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
#Initializing parameters
m=0.3
c=4
```

```python
#creating input and output
X=np.array([1,3,5,6,8,9,13,15,22,35,46])
Y=X+5
Y
```

```python
def cal_pred(x,m,c):
  return m*x+c
```

```python
#finding predicted outputs 
Y_pred=np.array([])
for i in X:
  y_hat=cal_pred(i,m,c)
  Y_pred=np.append(Y_pred,y_hat)
Y_pred
```

```python
#Plotting the given input against expected output
plt.plot(X[:],Y[:],c="red",label='Expected')
plt.plot(X[:],Y_pred[:],c="blue",label='Predicted')
plt.figure(figsize=(1,1))
plt.show()
```

```python
errors=np.array([])
def get_error(y_pred,y_hat):
  return (y_hat-y_pred)*(y_hat-y_pred)/Y.size
for i in range(11):
  err=get_error(Y_pred[i],Y[i])
  errors=np.append(errors,err)
errors
```

```python
#plotting error
plt.plot(errors)
plt.show()
```

```python
#FINDING MIN ERROR POINT IN WEIGHT
def slope_error(slope,intercept):
  arr=np.array([])
  for i in range(X.size):
      y_hat=slope*X[i]+intercept
      arr=np.append(arr,y_hat)
  MSE=mean_squared_error(Y,arr)
  return MSE
slope=[i/100 for i in range(0,95)]
cost=np.array([])
for i in slope:
  val=slope_error(i,c)
  cost=np.append(cost,val)
cost
```

```python
plt.plot(slope[:],cost[:])
plt.show()
```
